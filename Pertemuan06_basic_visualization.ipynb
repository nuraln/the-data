### Question: Compare the trend of top 5 countries that distributed the most immigration to Canada

```python
df.sort_values(by='Total', ascending=False, axis=0, inplace=True)

df_top5 = df.head(5)

df_top5 = df_top5[years].transpose()

df_top5.index = df_top5.index.map(int)

df_top5.plot(kind='line', figsize=(14, 8))

plt.title('Immigration trend of top 5 Countries')
plt.ylabel('Number of imigration')
plt.xlabel('Years')

plt.show();
```

## Area plots

```python
# Ubah tipe data ke integer
df_top5.index = df_top5.index.map(int)

df_top5.plot(kind='area', alpha=0.20, stacked=False , figsize=(20, 10))

plt.title('Immigration trend of top 5 Countries')
plt.ylabel('Number of imigration')
plt.xlabel('Years')

plt.show();
```

## Two tipes of plotting

Ada dua gaya dalam membuat plot di matplotliba.
* 1. Scripting layer (procedural method), menggunakan matplotlib.pyplot as plt
* 2. Artist layer (Object oriented method), using axes

```python
# option 2, object oriented
ax = df_top5.plot(kind='area', alpha=0.25, figsize=(20,10));

ax.set_title('Immigration trend of top 5 Countries')
ax.set_ylabel('Number of imigration')
ax.set_xlabel('Years')
```

### question: what the imigration distribution for Denmark, Norway and Sweden for years 1980 - 2013

```python
df.loc[['Denmark', 'Norway', 'Sweden'], years]
```

```python
# generate histogram
df.loc[['Denmark', 'Norway', 'Sweden'], years].plot.hist()
```

Kelihatannya tidak benar, malah yang di plot adalah distribusi frekuensi untuk years.

Kita bisa perbaiki dengan mentranspose dataset

```python
df_t = df.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_t.head()
```

```python
df_t.plot(kind='hist', figsize=(20, 6))
```

```python
# Make the plot esthetic

# Get x-tick values
count, bin_edges = np.histogram(df_t, 15)

df_t.plot(kind='hist', figsize=(20, 6), bins=15, alpha=0.6, xticks=bin_edges,
                                color=['coral', 'darkslateblue', 'mediumseagreen'] )

plt.title('Histogram of imigration from Denmark, Norway and Sweden')
plt.ylabel('Number of years')
plt.xlabel('Number of imigrants')
```

Jika tidak ingin plots saling tumpang tindih, kita dapat menumpuknya dengan parameter stacked. 

```python
count, bin_edges = np.histogram(df_t, 15)
xmin = bin_edges[0] - 10
xmax = bin_edges[-1] + 10
df_t.plot(kind='hist', 
          figsize=(20, 6), 
          bins=15, 
          alpha=0.6, xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen'],
          stacked=True,
          xlim=(xmin, xmax))

plt.title('Histogram of imigration from Denmark, Norway and Sweden')
plt.ylabel('Number of years')
plt.xlabel('Number of imigrants')
```

### Pie Chart

```python
df_continents = df.groupby('Continent', axis=0).sum()
df_continents
```

```python
df_continents['Total'].plot(kind='pie',
                           figsize=(5, 6),
                           autopct='%1.1f%%',
                           startangle=90,
                           shadow=True)

plt.title('Immigration to Canada')
plt.axis('equal')

plt.show();
```

```python
color_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]

df_continents['Total'].plot(kind='pie',
                           figsize=(15, 6),
                           autopct='%1.1f%%',
                           startangle=90,
                           shadow=True,
                           labels=None,
                           pctdistance=1.12,
                           colors=color_list,
                           explode=explode_list)
plt.title('Immigration to Canada', y=1.12)
plt.axis('equal')

# add legend
plt.legend(labels=df_continents.index, loc='upper left')
plt.show();
```
